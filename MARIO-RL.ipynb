##AI Performs Random actions without any logic

```python
env = gym.make("ALE/MarioBros-v5",render_mode = "rgb_array")
```

```python
num_episodes = 5
frames = []

for _ in range(num_episodes):
  env.reset(seed = 32)

  done = False
  while not done:
    action = env.action_space.sample()
    obs,reward,terminated,truncated,info = env.step(action)
    frames.append(obs)
    done = terminated or truncated

env.close()

clip = ImageSequenceClip(frames,fps = 30)
clip.write_videofile("./mario bros/random_actions.mp4",codec = "libx264")
```

```python
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
```

```python
def wrap():
  return gym.make("ALE/MarioBros-v5",render_mode = "rgb_array")

#simulated multiple enviroments out of one
env = DummyVecEnv([wrap])
```

```python
model = PPO("CnnPolicy",env)
```

```python
model.learn(1e4)
```

```python
model.save("/content/mario bros/ppo_model")
```

```python
model = PPO.load("/content/mario bros/ppo_model.zip")
```

```python
env = gym.make("ALE/MarioBros-v5",render_mode = "rgb_array")
frames = []

num_episodes = 5

for _ in range(num_episodes):
  state,info = env.reset()
  done = False
  while not done:
    action,_ = model.predict(state,deterministic=True)
    state,reward,terminated,truncated,info = env.step(action)
    frames.append(state)
    done = terminated or truncated

env.close()
```

#Using Custom Enviroment

```python
from gymnasium import Wrapper
```

```python
class Myenv(Wrapper):
  def __init__(self,env):
    super().__init__(env)
    self.prev_value = 0

  def reset(self,**kwargs):
    state,info = self.env.reset(**kwargs)
    self.prev_value = 0
    return state, info

  def step(self,action):
    state,reward,terminated,truncated,info = self.env.step(action)
    current_score = info.get("score",0)

    if current_score > self.prev_value:
      reward +=5

    self.prev_value = current_score
    return state,reward,terminated,truncated,info
```

```python
def wraper():
  env = gym.make("ALE/MarioBros-v5",render_mode = "rgb_array")
  env = Myenv(env)
  return env
env = DummyVecEnv([wraper])
```

```python
model = PPO("CnnPolicy",env)
```

```python
model.learn(1e4)
model.save("/content/mario bros/wraper_model")
```

```python
env = gym.make("ALE/MarioBros-v5",render_mode = "rgb_array")
frames = []
for _ in range(num_episodes):
  state,info = env.reset()
  done = False

  while not done:
    action,_ = model.predict(state,deterministic=True)

    state,reward,terminated,truncated,info = env.step(action)

    frames.append(state)

    done = truncated or terminated

env.close()
clip = ImageSequenceClip(frames,fps = 30)
clip.write_videofile("/content/mario bros/wrappper_actions.mp4")

Video("/content/mario bros/wrappper_actions.mp4",embed = True)

video = VideoFileClip("/content/mario bros/wrappper_actions.mp4")

duration = video.duration
min = int(duration//60)
seconds = int(duration%60)

print(f"The video duration is {min} minutes and {seconds} seconds")
```
